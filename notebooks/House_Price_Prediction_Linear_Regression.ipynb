The objective of this project is to preprocess a housing dataset, build a Linear Regression model to predict house prices, and evaluate the model using standard machine learning evaluation metrics such as Mean Squared Error (MSE) and R-squared (R²).

Machine learning projects require different Python libraries for handling data, performing mathematical computations, preprocessing datasets, training models, and evaluating results.

```python
import pandas as pd
import numpy as np
```

# 2. Loading the Dataset

The housing dataset is loaded into a Pandas DataFrame. Each row represents a house, while each column represents one characteristic (feature) of that house.

These features will later be used to train a machine learning model capable of predicting house prices.

```python
df = pd.read_csv(
    "../datasets/house Prediction Data Set.csv",
    sep=r"\s+",
    header=None
)
```

## Assigning Column Names

The original dataset does not include descriptive column names. Therefore, meaningful names are assigned to each column to improve readability and make the dataset easier to understand during analysis.

```python
df.columns = [
    "CRIM", "ZN", "INDUS", "CHAS", "NOX",
    "RM", "AGE", "DIS", "RAD", "TAX",
    "PTRATIO", "B", "LSTAT", "MEDV"
]
```

# 8. Feature and Target Selection

Machine learning models require two main components:

- **Features (X):** The input variables used to make predictions.
- **Target (y):** The output variable that the model aims to predict.

In this project, the target variable is **MEDV**, which represents the median value of owner-occupied homes.

```python
X = df.drop("MEDV", axis=1)
y = df["MEDV"]
```

### Observation

The **MEDV** column was selected as the target variable because it contains the house prices to be predicted.

The remaining 13 columns were used as input features for model training.

# 9. Feature Scaling

The features in the dataset have different numerical ranges. For example, **CRIM** contains relatively small values, while **TAX** contains much larger values.

To ensure that all features contribute fairly during model training, the data is standardized using **StandardScaler**.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

### Observation

Feature scaling transformed all input variables to a common scale with a mean of approximately **0** and a standard deviation of **1**.

This helps improve model performance and prevents features with larger values from dominating the learning process.

# 10. Splitting the Dataset

To evaluate the performance of the model fairly, the dataset is divided into two parts:

- **Training Set (80%)** – Used to train the model.
- **Testing Set (20%)** – Used to evaluate the model on unseen data.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)
```

### Observation

The dataset was split into training and testing sets using an **80:20 ratio**.

Using separate datasets for training and testing helps evaluate how well the model generalizes to unseen data and reduces the risk of overfitting.

# 11. Training the Linear Regression Model

A Linear Regression model is created and trained using the training dataset.

During training, the model learns the relationship between the input features and the target variable (house price).

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)
```

### Observation

The model was successfully trained using the training dataset and is now ready to make predictions on unseen data.

# 12. Making Predictions

After training the Linear Regression model, it is used to predict house prices for the testing dataset.

The predicted values are then compared with the actual house prices.

```python
y_pred = model.predict(X_test)
```

### Comparing Actual and Predicted Values

The following table displays the actual house prices alongside the values predicted by the model.

```python
comparison = pd.DataFrame({
    "Actual Price": y_test.values,
    "Predicted Price": y_pred
})

comparison.head(10)
```

# 13. Model Evaluation

To measure the performance of the Linear Regression model, two evaluation metrics are used:

- **Mean Squared Error (MSE):** Measures the average squared difference between the actual and predicted values.
- **R-squared (R²):** Indicates how well the model explains the variation in the target variable.

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)

r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

### Observation

The Linear Regression model achieved:

- **Mean Squared Error (MSE): 24.29**
- **R-squared (R²): 0.669**

The R² score indicates that the model explains approximately **67%** of the variation in house prices, demonstrating a reasonably good predictive performance for a basic Linear Regression model.

# 14. Interpreting Model Coefficients

The coefficients of the Linear Regression model indicate how each feature influences the predicted house price.

- A **positive coefficient** means that an increase in the feature tends to increase the predicted house price.
- A **negative coefficient** means that an increase in the feature tends to decrease the predicted house price.

```python
coefficients = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": model.coef_
})

coefficients
```

### Observation

The model coefficients show the influence of each feature on house prices.

For example:

- **RM (Number of Rooms)** has a positive coefficient, indicating that houses with more rooms generally have higher predicted prices.
- **CRIM (Crime Rate)** has a negative coefficient, suggesting that higher crime rates are associated with lower house prices.
- **NOX (Nitric Oxide Concentration)** also has a negative coefficient, indicating that higher pollution levels are associated with lower house prices.
